```python
s = np.random.uniform(0,1,100)
```

```python
count, bins, ignored = plt.hist(s, 20, density=True)
```

```python
def exponential_dis(x, alpha):
    x[x < 0] = 0
    x[x >= 0] = 1 - np.exp(- alpha * x)
    return x
```

```python
alpha = 2
l = np.random.uniform(-5,5,20)
k = exponential_dis(l, alpha)
```

```python
def inv_ex_dis(y, alpha):
    x = - (np.log(1 - y)) / alpha
    return x
```

```python
res  = inv_ex_dis(s, alpha)
```

```python
count, bins, ignored = plt.hist(res, 40, density=True)
```
